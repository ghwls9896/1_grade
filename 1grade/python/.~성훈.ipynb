```python
import numpy as np
import scipy.stats as sp
import matplotlib.pyplot as plt
n = 10
p=0.3
rv =sp.binom(n,p)

xx=np.arange(n+1)

plt.bar(xx,rv.cdf(xx))
plt.show()
```

```python
np.random.seed(0)
x=rv.rvs(30)
print(x)
```

```python
mean=10
var=9
rv=sp.norm(mean,var)
xx=np.arange(-2,22,0.1)
plt.bar(xx,rv.pdf(xx))
plt.show()
```

```python
np.random.seed(0)
x=rv.rvs(30)
print(x)
print(np.mean(x)) #평균
print(np.var(x)) #분산
```
