```python
import random
import time
import copy
from operator import itemgetter

NCARDS = 8
INIT_SCORE = 100
SUIT_TUPLE = ('Clubs', 'Hearts', 'Diamonds', 'Spades') # 값이 큰 순서대로 배치
RANK_TUPLE = ('Ace', '2', '3', '4', '5', '6', '7', '8', '9', '10', 'Jack', 'Queen', 'King')

def createDeck():
    deck = []
    for suit in SUIT_TUPLE:
        for value, rank in enumerate(RANK_TUPLE):
            for value_s ,suit in enumerate(SUIT_TUPLE):
                card = {'suit': suit, 'rank': rank, 'value': value+1, 'value_s' : value_s+1}
                deck.append(card)
    return deck

def shuffleCards(deck):
    random.shuffle(deck)
    return deck

def getCards(deck, numCards):
    cardList = []
    for i in range(numCards):
        card = deck.pop() # 맨 뒤에서 한장 제거
        cardList.append(card)
    return cardList

def my_addCard(deck, myCardList):
    card = deck.pop() 
    myCardList.append(card)
    return myCardList
    
def deal_addCard(deck, dealCardList):
    card = deck.pop() 
    dealCardList.append(card)
    return dealCardList


def sortCards_s(cards):
    cards.sort(key=itemgetter('value_s'), reverse=True)
    return cards

def sortCards(cards):
    cards.sort(key=itemgetter('value'), reverse=True)
    return cards
        
def isFlush(cards):
    
    suit = cards[0]['suit']
    isFlush = True
    
    for card in cards:
        if card['suit'] != suit:
            isFlush = False
            break
            
    if isFlush:
        return cards[0]['value']
    else:
        return None
    
def isStraight(cards):
    value_1 = cards[0]['value']
    value_2 = cards[1]['value']
    value_3 = cards[2]['value']
    value_4 = cards[3]['value']
    value_5 = cards[4]['value']
    isStraight = False
    
    if (value_4 == value_5+1):
        if(value_3 == value_4+1):
            if(value_2 == value_3+1):
                if(value_1 == value_2+1):
                    isStraight = True
            
    if isStraight:
        return value_1
    else:
        return None
    
def is4Cards(cards):
    value = cards[0]['value']
    value_1 = cards[0]['value']
    value_2 = cards[1]['value']
    value_3 = cards[2]['value']
    value_4 = cards[3]['value']
    value_5 = cards[4]['value']
    is4Cards = True
    if (value_1 != value_2):
        value = cards[1]['value']
        if (value_2 != value_3):
            is4Cards = False
        elif (value_4 != value_5):
            value = cards[4]['value']
            is4Cards = False
    else:
        if (value_2 != value_3):
            is4Cards = False
        elif(value_3 != value_4):
            is4Cards = False
        
    if is4Cards:
        return value
    else:
        return None
    
def isTriple(cards):
    value_1 = cards[0]['value']
    value_2 = cards[1]['value']
    value_3 = cards[2]['value']
    value_4 = cards[3]['value']
    value_5 = cards[4]['value']
    isTriple = False
    if(value_1 == value_2 and value_1 == value_3):
        value = cards[0]['value']
        isTriple = True
    if(value_2 == value_3 and value_2 == value_4):
        value = cards[1]['value']
        isTriple = True
    if(value_3 == value_4 and value_3 ==value_5):
        value = cards[2]['value']
        isTriple = True
    if isTriple:
        return value
    else:
        return None
    
def is2Pairs(cards):
    value_1 = cards[0]['value']
    value_2 = cards[1]['value']
    value_3 = cards[2]['value']
    value_4 = cards[3]['value']
    value_5 = cards[4]['value']
    is2Pairs = False
    if(value_1 == value_2):
        value = cards[0]['value']
        if (value_3 == value_4):
            is2Pairs = True
        if (value_4 == value_5):
            is2Pairs = True
    if(value_2 == value_3):
        value = cards[1]['value']
        if (value_4 == value_5):
            is2Pairs = True
                        
    if is2Pairs:
        return value
    else:
        return None
    
    
def is1Pairs(cards):
    value_1 = cards[0]['value']
    value_2 = cards[1]['value']
    value_3 = cards[2]['value']
    value_4 = cards[3]['value']
    value_5 = cards[4]['value']
    is1Pairs = False
    if(value_1 == value_2):
        value = cards[0]['value']
        is1Pairs = True
    if(value_2 == value_3):
        value = cards[1]['value']
        is1Pairs = True
    if(value_3 == value_4):
        value = cards[2]['value']
        is1Pairs = True
    if(value_4 == value_5):
        value = cards[3]['value']
        is1Pairs = True
            
            
    if is1Pairs:
        return value
    else:
        return None
    
    
def isHigh(cards):
    value = cards[0]['value']
    
    return value

def time_s():
    for i in range(3):
        time.sleep(0.5)
        print(">", end="")
        
def victory():
    print("당신의 승리입니다!")
    

def fail():
    print("당신의 패배입니다")
    
#def draw():
    
def print_List(cards):
    sign_cards = copy.deepcopy(cards)
    for i in range (len(sign_cards)):
        sign_cards[i].pop('value')
        sign_cards[i].pop('value_s')
        for v in sign_cards[i].values():
            print(v, end=" ")
        print(" ")
    return sign_cards


# def dis_Flush(myCardList, dealCardList):
#     if(type(isFlush(myCardList))== int and type(isFlush(dealCardList))==int):  #2명 모두 플러쉬일 경우
#         if(myCardList[]['value_s'])
#                                         if (isFlush(myCardList)>isFlush(dealCardList)):

#                                         elif (isFlush(myCardList)<isFlush(dealCardList)):

#                                         else:
#                                     elif (type(isFlush(myCardList))==int and type(isFlush(dealCardList))!=int):

#                                     elif (type(isFlush(myCardList))!= int and type(isFlush(dealCardList))==int):

#                                     else:
#                                         break




        
def game_poker():
    score = 0
    dealscore =0
    choice = None
    repeat_count = 0
    time_s()
    print("덱을 생성중입니다.")
    time.sleep(1)
    deck = createDeck() ##덱 생성
    time_s()
    print("덱을 섞는중입니다.")
    time.sleep(1)
    deck = shuffleCards(deck) ## 덱 섞기
    time_s()
    print("당신과 딜러가 서로 2장의 카드를 받고 그것을 공개합니다.")
    time.sleep(1)
    myCardList = getCards(deck, 2)
    dealCardList = getCards(deck, 2)
    myCardList = sortCards_s(myCardList)  # value값에 순서대로 정렬
    dealCardList = sortCards_s(dealCardList) # value값에 순서대로 정렬
    time_s()
    print("내가 받은 카드")
    print_List(myCardList)
    print("\n\n")
    time_s()
    print("딜러가 받은 카드")
    print_List(dealCardList)
    print("")

    while((choice != 'd') or (choice != 'g')):
        time_s()
        choice = input("계속 진행하시려면 GO,기권하시려면 DROP (g/d)\n")
        if(choice == 'd'):
            score -= (10*(repeat_count+1))
            print(f">>>당신의 점수는 {score}점 입니다.")
            return (score, dealscore)
        elif(choice == 'g'):
            time_s()
            print("당신과 딜러가 한 장의 카드를 더 받습니다")
            my_addCard(deck, myCardList)
            deal_addCard(deck, dealCardList)
            myCardList = sortCards_s(myCardList)
            dealCardList = sortCards_s(dealCardList)
            time_s()
            print("내 카드")
            print_List(myCardList) # value값에 순서대로 정렬
            if (len(myCardList)==5):
                break
        else:
            print(">>> [ERROR] g/d중에 하나를 정확히 입력하세요!!!")
            repeat_count -= 1
        repeat_count += 1

    #if (len(myCardList)==5):
        
        
                                    

                                    
                                    
                                    
```
