```python
import numpy as np
import scipy. stats as sp
import matplotlib.pyplot as plt
# 문제 X-N(10,9)
# 1x의 pdf를 그려라
# 2x의 확률 분포를 따르는 표본 30개를 만들어 평균과 분산을 계산하라
# 3 이 표본 30개를 50회 생성한 후 평균과 분산을 계산
```

```python
mean = 10 #평균
var = 9  #분산
rv =sp.norm(mean,var**(1/2))
xx=np.arange(-2,22,0.1)
plt.bar(xx,rv.pdf(xx))

plt.show()
```

```python
#1. X 의 확률 분포를 생성하는 코드를 작성하고, 이 학률 분포의 누적 분포 함수를 그려라. 
import numpy as np
import scipy. stats as sp
import matplotlib.pyplot as plt
n= 50
p =0.6
rv =sp.binom(n,p)

xx=np.arange(n+1)

plt.bar(xx,rv.cdf(xx))
plt.show()
```

```python
#2. X 의 확률 분포를 따르는 100개의 표본을 생성하고, 각 표본의 평균을 구하라. 
np.random.seed(0)
x = rv.rvs(100)
print(np.mean(x))
```
