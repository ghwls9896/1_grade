```python
%%writefile lee_ho_jin
def header_text(width, text):
    print("*", end="")
    lspace = int((width-len(text))/2)
    rspace = int(width - len(text)-lspace)
    print_space(lspace)
    print(text, end="")
    print_space(rspace)
    print("*")
def print_space(count):
    for i in range(count):
        print(" ", end="")
def star_line(count):
    for i in range(count):
        print("*", end="")
    print()
def header_print(title, data , name):
    line_len = max(len(title), len(data), len(name))
    star_line(line_len+4)
    header_text(line_len+2,title)
    header_text(line_len+2,data)
    header_text(line_len+2,name)
    star_line(line_len+4)
```

```python
import math
def print_sin(stat_deg, end_deg,length,width) :
    num_width = 0
    num_height = 0
    midd = width/2
    width += 1          
    print(str("x").rjust(2), end=" ")
    for num_width in range (width):
        print("-", end="")
    print(str("sin(x)").ljust(2))
    
    for num_height in range (stat_deg,length):
        x_val = num_height*10
        sin_val = (round(math.sin(math.pi * (x_val / end_deg)), 3))
        r_loc = round(midd + (midd * sin_val))
        print(str(num_height).zfill(2), end="0")
        print("", end="")
        for num_width in range (width):
            print(" ", end ="")
            if(num_height == 0 or sin_val == 0):
                if (num_width == midd):
                    print("*", end="")

            else:
                if (num_width == midd-1):
                    print("|", end="")
                        
                if(num_width == r_loc):
                    print("*", end="")
            if (num_width == width-1):
                print(str(sin_val).zfill(2), end="0")
                print("")     
```
