```python
import random, sys
import keyboard, time
PLAYER_X    = ' ●'
PLAYER_O    = ' ○'
BOARD_WIDTH  = 19
BOARD_HEIGHT = 19
EMPTY_SPACE = ' . '
```

```python
def getNewBoard():
    board = {}

    for row in range(BOARD_HEIGHT):
        for col in range(BOARD_WIDTH):
            board[(col, row)] = EMPTY_SPACE
    return board


def isFull(board):
    
    for row in range(BOARD_HEIGHT):
        for col in range(BOARD_WIDTH):
            if board[(col, row)] == EMPTY_SPACE:
                return False
    return True


def getKey(board,row, col, flag):
    while True:
        if keyboard.is_pressed('down'):
            next_row = row
            next_col = col+1
            if (board[(next_row,next_col)]!=EMPTY_SPACE):
                print('이미 컬럼에 알이 들어가있습니다')
                return (row,col,flag)
            else:
                board[(next_row,next_col)] = board[(row, col)]
                board[(row,col)] = EMPTY_SPACE
                displayBoard(board)
                time.sleep(0.2)
                return (next_row, next_col,flag)
        if keyboard.is_pressed('up'):
            next_row = row
            next_col = col-1
            if (board[(next_row,next_col)]!=EMPTY_SPACE):
                print('이미 컬럼에 알이 들어가있습니다')
                return (row,col, flag)
            else:
                board[(next_row,next_col)] = board[(row, col)]
                board[(row,col)] = EMPTY_SPACE
                displayBoard(board)
                time.sleep(0.2)
                return (next_row, next_col,flag)
        if keyboard.is_pressed('left'):
            next_row = row-1
            next_col = col
            if (board[(next_row,next_col)]!=EMPTY_SPACE):
                print('이미 컬럼에 알이 들어가있습니다')
                return (row,col,flag)
            else:
                board[(next_row,next_col)] = board[(row, col)]
                board[(row,col)] = EMPTY_SPACE
                displayBoard(board)
                time.sleep(0.2)
                return (next_row, next_col,flag)
        if keyboard.is_pressed('right'):
            next_row = row+1
            next_col = col
            if (board[(next_row,next_col)]!=EMPTY_SPACE):
                print('이미 컬럼에 알이 들어가있습니다')
                return (row,col, flag)
            else:
                board[(next_row,next_col)] = board[(row, col)]
                board[(row,col)] = EMPTY_SPACE
                displayBoard(board)
                time.sleep(0.2)
                return (next_row, next_col,flag)
        if keyboard.is_pressed('esc'):
            print('종료 키 입력')
            displayBoard(board)
            time.sleep(0.2)
            break

        if keyboard.is_pressed('enter'):
            print(f'돌을({row},{col})자리에 두었습니다! ')
            displayBoard(board)
            print(f"({row}, {col})")
            time.sleep(0.2)
            flag = 0
            return (row,col,flag)
        
        
def askForPlayerChoice(playerTile, board):
    next_row = 9
    next_col = 9
    flag = 1
    board[(next_row, next_col)] = playerTile
    while (flag == 1):
        (next_row, next_col, flag) = getKey(board, next_row, next_col, flag)
            
        
        
def displayBoard(board):
    tileChars = [] # 1차원 리스트

    # board -> tileChars에 복사
    for row in range(BOARD_HEIGHT):
        for col in range(BOARD_WIDTH):
            tileChars.append(board[(col, row)])
            
    # 화면에 tileChars 출력하기
    print("""
    +  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  +
    | {}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{} |
    | {}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{} |
    | {}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{} |
    | {}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{} |
    | {}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{} |
    | {}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{} |
    | {}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{} |
    | {}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{} |
    | {}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{} |
    | {}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{} |
    | {}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{} |
    | {}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{} |
    | {}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{} |
    | {}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{} |
    | {}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{} |
    | {}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{} |
    | {}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{} |
    | {}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{} |
    | {}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{}{} |
    +  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  -  +""".format(*tileChars))
```
